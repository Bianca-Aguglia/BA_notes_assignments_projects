## Colors

```python
fig, ax = plt.subplots()
x = np.linspace(0,10,1000)
y = 2*x
# set of default color and style
ax.plot(x, np.cos(x))
# RGB tuple, values 0 to 1, solid style
ax.plot(x, np.sin(x), color=(1.0,0.2,0.3),linestyle='-')
# specify color by name, dashed style
ax.plot(x, y, color='blue', linestyle='--')
# short color code (rgbcmyk), dotted style   
ax.plot(x, x+3, color='g',linestyle=':')
# Grayscale between 0 and 1, dashdot style    
ax.plot(x, np.log(x), color='0.75',linestyle='-.')
# Hex code (RRGGBB from 00 to FF), dashed style  
ax.plot(x, x, color='#FFDD44',linestyle='--')
```
