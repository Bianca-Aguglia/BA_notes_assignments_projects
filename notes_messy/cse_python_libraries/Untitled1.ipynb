```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100, endpoint = True)
y = np.random.randint(-5, 5, 100)
```

```python
fig, ax = plt.subplots(figsize = (6,6))

# rings
n = 50
min_size = 50
max_size = 50*50

# ring positions
P = np.random.uniform(0, 1, (n, 2))

# ring colors
C = np.ones((n,4)) * (0,0,0,1)
C[:,3] = np.linspace(0,1,n)

# ring sizes
S = np.linspace(min_size, max_size, n)

ax.scatter(P[:,0], P[:,1], s = S, edgecolors = C, facecolors = 'None')
ax.set_xlim(0,1)
ax.set_xticks([])
ax.set_ylim(0,1)
ax.set_yticks([])

```

```python
plt.axes([0.1, 0.1, 0.3, 0.3], facecolor = 'gray')
plt.axes([0.5, 0.5, 0.7, 0.8], facecolor = 'blue')
plt.plot(x,y)
```

```python
fig = plt.figure(figsize = (10,8), facecolor = 'green', dpi = 40)
ax1 = plt.axes([0.1, .1, .5, .5])
ax2 = plt.axes([0.7, 0.7, 0.9, 0.9])
ax1.plot(x,y)
```
