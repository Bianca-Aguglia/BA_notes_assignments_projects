```python
def odd():
    funcs = []
    for c in 'abcd':
        funcs.append(lambda: c)
    return funcs
```

```python
for func in odd():
    print(func(), end=' ')
```

```python
def odd1():
    funcs = []
    for c in 'abcd':
        funcs.append(lambda c=c: c)
    return funcs
```

```python
for func in odd1():
    print(func(), end = ' ')
```
